**Problem 1: Unique Words in a Sentence**
Logic:

1)Take the input sentence as a string.

2)Split the sentence into individual words using the split() method.

3)Convert the list of words into a set to automatically remove duplicate words.

4)Count the number of unique words using len().

5)Print both the count and the set of unique words.

```python
sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

**Problem 2: Highest Salary from Employee Data**
Logic

Store employee names and salaries in a dictionary where:

Key â†’ Employee name

Value â†’ Salary

Use the max() function with dictionary.get to find the key having the highest salary value.

Display the employee name and corresponding highest salary.

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

**Problem 3: Find Maximum and Minimum Value** Logic

Initialize both maximum and minimum values with the first element of the list.

Traverse the list using a loop.

Compare each element:

If current value is greater than maximum, update maximum.

If current value is less than minimum, update minimum.

Print the final maximum and minimum values.

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

**Problem 4: Count Products Above a Price Threshold** Logic

Store product prices in a list.

Initialize a counter variable to zero.

Loop through each price in the list.

If the price is greater than 1000, increment the counter.

Print the total count of products above the threshold.

```python
prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

**Problem 7: Count Character Frequency** Logic

Take the input string.

Create an empty dictionary to store character frequencies.

Traverse each character in the string.

If the character already exists in the dictionary, increment its count.

If not, add the character to the dictionary with a count of 1.

Print the dictionary containing character frequencies.

```python
text = "python"
char_frequency = {}

for char in text:
    char_frequency[char] = char_frequency.get(char, 0) + 1

print(char_frequency)
```
