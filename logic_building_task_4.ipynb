**Problem 1: Mobile Recharge Validation System**


Loops


Recharge amount must be ₹50 or more

Recharge amount must match valid plans

Allow retry until valid input is given

Use a function for validation

Use a while loop for retry

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount >= 50 and amount in valid_plans:
        return True
    else:
        return False


# Retry until valid recharge
while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge Successful")
        break
    else:
        print("Invalid recharge amount. Please try again.")
```

**Problem 2: Inventory Reorder Alert System**


 Logic



Store product stock in a dictionary

Loop through dictionary items

If stock < 15 → print Reorder Alert

Otherwise → print Stock OK

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

check_inventory(inventory)
```

**Problem 3: Student Result Processing System**


Logic


Accept a list of marks

Calculate total and average

If average ≥ 50 → Pass

Else → Fail

Return result from function

```python
def student_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks = [55, 60, 45, 70, 50]
result = student_result(marks)

print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**


 Logic



Base fare = ₹50

₹12 per km

Peak hour → add 25%

Use function to calculate fare

Use while loop for retry

```python
def calculate_fare(distance, peak_hour):
    fare = 50 + (distance * 12)

    if peak_hour:
        fare += fare * 0.25  # 25% extra

    return fare


while True:
    distance = int(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    peak_hour = True if peak == "yes" else False

    total_fare = calculate_fare(distance, peak_hour)
    print("Total Fare: ₹", int(total_fare))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**


 Logic



Count present days ("P")

Calculate attendance percentage

If ≥ 75% → Eligible

Else → Not Eligible

Return result from function

```python
def attendance_eligibility(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance = ["P", "P", "A", "P", "P", "A", "P"]
result = attendance_eligibility(attendance)

print("Attendance Status:", result)
```

**Problem 6: Password Strength Checker**


 Logic



Minimum length = 8

Must contain:

At least one digit

At least one special character (@ # $)

Use loop to check characters

Return strength result

```python
def check_password_strength(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter password: ")
print(check_password_strength(password))
```
