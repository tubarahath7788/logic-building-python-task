**Problem 1: Smart Parking Lot Management System**
 Logic

1)Initialize a variable to keep track of currently parked vehicles.

2)Loop through the vehicle entry and exit logs.

3)Increase the count when a vehicle enters ("IN").

4)Decrease the count when a vehicle exits ("OUT").

5)Compare the final parked count with the parking capacity.

6)Display parking status as Available or Full based on capacity.

 Key Concepts Used:

Function

List

Loop

Conditional statements

```python
# Function to manage smart parking lot
def parking_lot_manager(capacity, vehicle_logs):
    # Variable to track current parked vehicles
    parked = 0

    # Loop through each vehicle log
    for log in vehicle_logs:
        if log == "IN":
            parked += 1      # Vehicle enters parking
        elif log == "OUT":
            parked -= 1      # Vehicle exits parking

    # Display current parked vehicles
    print("Currently Parked Vehicles:", parked)

    # Check parking status based on capacity
    if parked > capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")


# Sample input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

# Function call
parking_lot_manager(capacity, logs)
```

**Problem 2: Online Food Delivery Time Estimator**
 Logic:

1)Calculate base delivery time using distanc(minutes per km).

2)Add extra time based on traffic conditions.

3)Add additional delay based on weather conditions.

4)Sum all delays to get the final estimated delivery time.

5)Display the final ETA.

 Key Concepts Used:

Function

Conditional logic

Mathematical calculations

```python
# Function to estimate food delivery time
def delivery_time_estimator(distance, traffic, weather):
    # Base delivery time: 5 minutes per km
    time = distance * 5

    # Add delay based on traffic condition
    if traffic == "High":
        time += 15
    elif traffic == "Medium":
        time += 10

    # Add delay based on weather condition
    if weather == "Rainy":
        time += 10
    elif weather == "Stormy":
        time += 20

    # Display estimated delivery time
    print("Estimated Delivery Time:", time, "minutes")


# Sample input
distance = 8
traffic_level = "High"
weather_condition = "Rainy"

# Function call
delivery_time_estimator(distance, traffic_level, weather_condition)
```

**Problem 3: Movie Theatre Seat Occupancy Analyzer**
 Logic:

1)Count the total number of booked seats.

2)Calculate occupancy percentage using:

(Booked Seats / Total Seats) × 100

Based on occupancy percentage:

≥ 90% → Housefull

≥ 70% → Almost Full

Else → Seats Available

3)Display occupancy percentage and show status.

 Key Concepts Used:

Function

List

Percentage calculation

Conditional checks

```python
# Function to analyze seat occupancy
def seat_occupancy(total_seats, booked_seats):
    # Count booked seats
    booked_count = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    # Display occupancy
    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy >= 90:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

# Function call
seat_occupancy(total_seats, booked_seats)
```

**Problem 4: Cloud Server Load Classification System**
 Logic:

1)Loop through CPU usage readings and calculate total usage.

2)Calculate average CPU load.

3)Classify server status:

Below 50% → Normal

50%–80% → Warning

Above 80% → Critical

4)Display average load and server status.

 Key Concepts Used:

Function

Loop

Average calculation

Conditional logic

```python
# Function to classify server load
def server_load(cpu_readings):
    # Calculate total CPU usage
    total = 0
    for cpu in cpu_readings:
        total += cpu

    # Calculate average CPU usage
    average = total / len(cpu_readings)

    # Display average CPU load
    print("Average CPU Load:", int(average), "%")

    # Classify server status
    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample input
cpu_data = [45, 60, 70, 85, 90]

# Function call
server_load(cpu_data)
```

# Problem 5: Smart Classroom Resource Usage Monitor
 Logic:

1)Loop through resource usage dictionary.

2)Check usage hours for each resource.

3)Identify resources used beyond the allowed threshold.

4)If any resource is overused, raise an energy alert.

5)Display overused resources and alert status.

 Key Concepts Used:

Function

Dictionary

Loop

Conditional statements

```python
# Function to monitor classroom resource usage
def resource_monitor(usage_data):
    # List to store overused resources
    overused = []

    # Loop through each resource and usage hours
    for resource, hours in usage_data.items():
        if hours > 8:   # Threshold for overuse
            overused.append(resource)

    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample input
resources = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

# Function call
resource_monitor(resources)
```

**Problem 6: Online Event Registration Capacity Controller**
 Logic:

1)Compare total registrations with event capacity.

2)If registrations exceed capacity:

  a)Confirm registrations up to capacity.

  b)Remaining users go to waitlist.

  c)Close registration.

3)If within capacity:

  a)Confirm all registrations.

  b)Keep registration open.

4)Display confirmed, waitlisted users, and registration status.

 Key Concepts Used:

Function

Conditional logic

Mathematical operations

```python
# Function to manage event registrations
def registration_controller(capacity, registrations):
    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    # Display registration details
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample input
event_capacity = 100
total_registrations = 105

# Function call
registration_controller(event_capacity, total_registrations)
```
