**Problem 1: Employee Performance Bonus Eligibility**

 Logic:

Store employee names and scores in a dictionary.

Find the highest score using max().

Iterate through the dictionary to find all employees with that score (handle ties).

Display their names along with the score.

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
highest_score = max(employees.values())

# Find employees with highest score
top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

**Problem 2: Search Query Keyword Analysis**


 Logic:

Convert the input sentence to lowercase.

Split the sentence into words.

Use a dictionary to count the frequency of each word.

Display only words that appear more than once.

```python
query = "Buy mobile phone buy phone online"

# Convert to lowercase and split into words
words = query.lower().split()

keyword_count = {}

for word in words:
    keyword_count[word] = keyword_count.get(word, 0) + 1

# Filter keywords appearing more than once
result = {word: count for word, count in keyword_count.items() if count > 1}

print(result)
```

**Problem 3: Sensor Data Validation**


 Logic:

Iterate through sensor readings using enumerate() to get index and value.

Check if the reading is even.

Store valid readings as (index, value) pairs.

Ignore odd readings.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**Problem 4: Email Domain Usage Analysis**

Logic:

Extract domain names from email IDs.

Count occurrences of each domain using a dictionary.

Calculate percentage usage based on total emails.

Display domain usage percentages.

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {int(percentage)}%")
```

**Problem 5: Sales Spike Detection**

 Logic:

Calculate average sales.

Determine threshold as 30% above average.

Iterate through sales list.

Display day number and sales value for spikes.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3

for index, value in enumerate(sales):
    if value > threshold:
        print(f"Day {index + 1}: {value}")
```

**Problem 6: Duplicate User ID Detection**

Logic:

Store user IDs in a list.

Use a dictionary to count occurrences.

Display only user IDs that appear more than once.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

user_count = {}

for user in user_ids:
    user_count[user] = user_count.get(user, 0) + 1

for user, count in user_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
