**Problem 1: Smart Light Controller**

**Logic:**

Create a class to represent a smart light.

Use a constructor to store the light name and set default status as OFF.

Create methods to turn the light ON and OFF.

Create another method to display the current status of the light.

Create an object of the class and perform the required action.

**OOPS Concepts Used:**

Class and Object

Constructor

Methods

```python
# Class to control a smart light
class SmartLight:
    # Constructor to initialize light name and status
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default light status is OFF

    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"

    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"

    # Method to display current light status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Creating an object of SmartLight class
light = SmartLight("Bedroom Light")

# Turning the light ON
light.turn_on()

# Displaying the light status
light.display_status()
```

**Problem 2: Employee ID Card System**

** Logic:**

Create a class to store employee details such as name, ID, and department.

Use a constructor to initialize employee information.

Create a method to display ID card details in a structured format.

Create an object of the class and call the display method.

**OOPS Concepts Used:**

Class

Constructor

Instance variables

Method calling

```python
# Class to store employee details
class EmployeeIDCard:
    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    # Method to display employee ID card
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Creating an object of EmployeeIDCard class
employee = EmployeeIDCard("Rahul", "EMP102", "IT")

# Displaying employee ID card
employee.display_id_card()
```

**Problem 3: Mobile Contact Record**

**Logic:**

Create a class to store contact name and phone number.

Initialize values using the constructor.

Create a method to display contact details.

Create an object of the class to save and display contact information.

**OOPS Concepts Used:**

Class and Object

Constructor

Methods

```python
# Class to store mobile contact details
class Contact:
    # Constructor to initialize contact name and phone number
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    # Method to display contact details
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Creating an object of Contact class
contact = Contact("Anita", 9876543210)

# Displaying saved contact
contact.display_contact()
```

**Problem 4: Product Price Tag Generator**

**Logic:**

Create a class to store product name and price.

Initialize product details using a constructor.

Create a method to display a formatted price tag.

Create an object and call the display method.

 **OOPS Concepts Used**:

Class

Constructor

Instance variables

Methods

```python
# Class to store product details
class Product:
    # Constructor to initialize product name and price
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    # Method to display product price tag
    def display_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Creating an object of Product class
product = Product("Headphones", 2499)

# Displaying product price tag
product.display_price_tag()
```

**Problem 5: Movie Rating Display System**

**Logic:**

Create a class to store movie name and rating.

Initialize the movie details using a constructor.

Create a method to display the movie rating in a readable format.

Create an object of the class and display the movie details.

 **OOPS Concepts Used**:

Class and Object

Constructor

Methods

```python
# Class to store movie details
class Movie:
    # Constructor to initialize movie name and rating
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    # Method to display movie rating
    def display_movie_details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")


# Creating an object of Movie class
movie = Movie("Inception", 4.8)

# Displaying movie details
movie.display_movie_details()
```

 **Problem 6: Delivery Address Manager**

 **Logic:**

Create a class to store customer name and delivery address.

Initialize values using a constructor.

Create a method to display delivery details.

Create an object of the class and call the display method.

**OOPS Concepts Used:**

Class

Constructor

Method calling

```python
# Class to store delivery details
class Delivery:
    # Constructor to initialize customer name and address
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    # Method to display delivery details
    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Creating an object of Delivery class
delivery = Delivery("Suman", "Hyderabad")

# Displaying delivery details
delivery.display_delivery_details()
```
